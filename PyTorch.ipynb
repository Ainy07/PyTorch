```python
import torch
import torch.nn as nn

X = torch.tensor([[1.0], [2.0], [3.0]])
Y = torch.tensor([[3.0], [6.0], [9.0]])

model = nn.Linear(1,1)

loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

for i in range(50):
    y_pred = model(X)
    loss = loss_fn(y_pred, Y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

test = torch.tensor([[10.0]])
print("Prediction:", model(test).item())
```
